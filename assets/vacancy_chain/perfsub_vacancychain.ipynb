```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.facecolor'] = 'white'
```

```python
def demand_function1(p1, a, b, m, p2):
  """Calculates the demand for good 1 given its price, utility function parameters, income, and price of good 2.

  Args:
    p1: Price of good 1
    a: Coefficient of good X in the utility function.
    b: Coefficient of good Y in the utility function.
    m: Income.
    p2: Price of good 2

  Returns:
    Demand for good 1
  """

  if p1 / p2 > a / b:
    return 0
  else:
    return m / p1
  
def demand_function2(p2, a, b, m, p1):
  """Calculates the demand for good 2 given its price, utility function parameters, income, and price of good 1.

  Args:
    p2: Price of good 2
    a: Coefficient of good X in the utility function.
    b: Coefficient of good Y in the utility function.
    m: Income.
    p1: Price of good 1

  Returns:
    Demand for good 2
  """

  if p1 / p2 < a / b:
    return 0
  else:
    return m / p2
  
# Linear supply function 

def supply_function(p,c1,c2):
  return c1+c2*p
```

```python
# n=3 case
# shift supply of x2

# u1(x1,x2) = x1 + 2x2
# u2(x1,x2) = x1 + 3x2
# u3(x1,x2) = x1 + 4x2
# m1=10, m2=20, m3=30

# equilibrium 1
# p1=3, p2=10
# x1=10, x2 = 3
# S1(p1) = -26 + 12 * p1
# S2(p2) = 3/10 * p2

# equilibrium 2: shift S2
# S2_new(p2) = 50/49 * p2
# p1=5/2, p2=7
# x1=4, x2 = 50/7
```

```python
# demand for x1
p1_range = np.linspace(0.5, 7, 1000)

d11_eqm1 = [demand_function1(p1,1,2,10,10) for p1 in p1_range]
d21_eqm1 = [demand_function1(p1,1,3,20,10) for p1 in p1_range]
d31_eqm1 = [demand_function1(p1,1,4,30,10) for p1 in p1_range]

d11_eqm2 = [demand_function1(p1,1,2,10,7) for p1 in p1_range]
d21_eqm2 = [demand_function1(p1,1,3,20,7) for p1 in p1_range]
d31_eqm2 = [demand_function1(p1,1,4,30,7) for p1 in p1_range]

D1_eqm1 = [x11+x21+x31 for x11,x21,x31 in zip(d11_eqm1,d21_eqm1,d31_eqm1)]
D1_eqm2 = [x11+x21+x31 for x11,x21,x31 in zip(d11_eqm2,d21_eqm2,d31_eqm2)]
S1_eqm1 = [supply_function(p1,-26,12) for p1 in p1_range]

```

```python
# individual demand curves: x1

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(d11_eqm1, p1_range, label = "Person 1")
plt.plot(d21_eqm1, p1_range, label = "Person 2")
plt.plot(d31_eqm1, p1_range, label = "Person 3")
plt.xlabel("Quantity of old apartments")
plt.ylabel("Price of old apartments")
plt.title("Individual demand curves: Old apartments")
plt.grid(True)
ax.set_xlim(0, 50)
plt.legend()
plt.show()
```

```python
# aggregate demand for x1

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D1_eqm1, p1_range, label="Demand")
plt.xlabel("Quantity of old apartments")
plt.ylabel("Price of old apartments")
plt.title("Aggregate demand curve: Old apartments")
plt.grid(True)
ax.set_xlim(0, 35)
plt.legend()

plt.annotate('Person 1 buys',xy=(2.8,4.5), xytext=(5.4, 5.2), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))
plt.annotate('Person 1 & 2 buy',xy=(11,2.8), xytext=(14,3.5), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))
plt.annotate('Person 1, 2, & 3 buy',xy=(26,2.4), xytext=(27,3.1), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# aggregate demand and supply for x1

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D1_eqm1, p1_range, label="Demand")
plt.plot(S1_eqm1, p1_range, label="Supply")
plt.xlabel("Quantity of old apartments")
plt.ylabel("Price of old apartments")
plt.title("Supply and demand equilibrium: Old apartments")
plt.grid(True)
ax.set_xlim(0, 30)
plt.legend()

plt.annotate('D',xy=(18,2.6), fontsize=15)
plt.annotate('S',xy=(16,3.65), fontsize=15)

plt.annotate('Equilibrium',xy=(10,3.05), xytext=(11,4.05), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# aggregate demand and supply for x1, eqm 2

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D1_eqm1, p1_range, label="Demand curve, initial")
plt.plot(D1_eqm2, p1_range, label="Demand curve, final")
plt.plot(S1_eqm1, p1_range, label="Supply curve")
plt.xlabel("Quantity of old apartments")
plt.ylabel("Price of old apartments")
plt.title("Effect of decreased price(new apartments) on old apartments")
plt.grid(True)
ax.set_xlim(0, 30)
plt.legend()

# demand shift arrows
plt.annotate('',xy=(22,1.85), xytext=(22, 2.45),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D1',xy=(22,2.6), fontsize=15)
plt.annotate('D2',xy=(22,1.5), fontsize=15)
plt.annotate('S',xy=(16,3.65), fontsize=15)

plt.annotate('Old equilibrium',xy=(10,3.05), xytext=(11,4.05), fontsize=14,
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8, alpha=0.4), alpha=0.4)

plt.annotate('New equilibrium',xy=(4,2.45), xytext=(0.5,1.05), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# demand for x2
p2_range = np.linspace(1, 15, 1000)

d12_eqm1 = [demand_function2(p2,1,2,10,3) for p2 in p2_range]
d22_eqm1 = [demand_function2(p2,1,3,20,3) for p2 in p2_range]
d32_eqm1 = [demand_function2(p2,1,4,30,3) for p2 in p2_range]

d12_eqm2 = [demand_function2(p2,1,2,10,5/2) for p2 in p2_range]
d22_eqm2 = [demand_function2(p2,1,3,20,5/2) for p2 in p2_range]
d32_eqm2 = [demand_function2(p2,1,4,30,5/2) for p2 in p2_range]

D2_eqm1 = [x12+x22+x32 for x12,x22,x32 in zip(d12_eqm1,d22_eqm1,d32_eqm1)]
D2_eqm2 = [x12+x22+x32 for x12,x22,x32 in zip(d12_eqm2,d22_eqm2,d32_eqm2)]
S2_eqm1 = [supply_function(p2,0,3/10) for p2 in p2_range]
S2_eqm2 = [supply_function(p2,0,50/49) for p2 in p2_range]
```

```python
# individual demand curves: x2

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(d12_eqm1, p2_range, label = "Person 1")
plt.plot(d22_eqm1, p2_range, label = "Person 2")
plt.plot(d32_eqm1, p2_range, label = "Person 3")
plt.xlabel("Quantity of new apartments")
plt.ylabel("Price of new apartments")
plt.title("Individual demand curves: New apartments")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()
plt.show()
```

```python
# aggregate demand for x2
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve")
plt.xlabel("Quantity of new apartments")
plt.ylabel("Price of new apartments")
plt.title("Aggregate demand curve: New apartments")
plt.grid(True)
ax.set_xlim(0, 20)
plt.legend()

plt.annotate('Person 3 buys',xy=(3.2,10), xytext=(6, 10.5), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))
plt.annotate('Person 3 & 2 buy',xy=(7,7.5), xytext=(8.5,8.2), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))
plt.annotate('Person 3, 2, & 1 buy',xy=(13,5), xytext=(15.5,6.2), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# aggregate supply and demand for x2
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve")
plt.plot(S2_eqm1, p2_range, label="Supply curve")
plt.xlabel("Quantity of new apartments")
plt.ylabel("Price of new apartments")
plt.title("Initial equilibrium: New apartments")
plt.grid(True)
ax.set_xlim(0, 20)
plt.legend()

# label curves
plt.annotate('D',xy=(11.5,5.3), fontsize=15)
plt.annotate('S',xy=(1.5,6.5), fontsize=15)

plt.annotate('Equilibrium',xy=(3.2,10), xytext=(6, 10.5), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# shift supply for x2

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve")
plt.plot(S2_eqm1, p2_range, label="Supply curve, initial")
plt.plot(S2_eqm2, p2_range, label="Supply curve, final")
plt.xlabel("Quantity of new apartments")
plt.ylabel("Price of new apartments")
plt.title("Shifting the supply of new apartments")
plt.grid(True)
ax.set_xlim(0, 20)
plt.legend()

# supply shift arrow
plt.annotate('',xy=(5.3,5.5), xytext=(2, 5.5),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D',xy=(11.5,5.3), fontsize=15)
plt.annotate('S1',xy=(1,6.5), fontsize=15)
plt.annotate('S2',xy=(5.7,5), fontsize=15)

plt.annotate('Old equilibrium',xy=(3.2,10), xytext=(6, 10.5), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8, alpha=0.4), alpha=0.4)
plt.annotate('New equilibrium',xy=(7.3,7), xytext=(10.1, 7.5), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8))

plt.show()
```

```python
# aggregate supply and demand for x2
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve, initial")
plt.plot(S2_eqm1, p2_range, label="Supply curve, initial", alpha=0.3)
plt.plot(S2_eqm2, p2_range, label="Supply curve, final")
plt.plot(D2_eqm2, p2_range, label="Demand curve, final")
plt.xlabel("Quantity of new apartments")
plt.ylabel("Price of new apartments")
plt.title("Effect of decreased price(old apartments) on demand for new apartments")
plt.grid(True)
ax.set_xlim(0, 20)
plt.legend()

# demand shift arrows
plt.annotate('',xy=(4.5,7.6), xytext=(4.5,8.9),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D1',xy=(4.2,9.2), fontsize=15)
plt.annotate('D2',xy=(4.2,6.9), fontsize=15)
plt.annotate('S1',xy=(1,6.5), fontsize=15, alpha=0.3)
plt.annotate('S2',xy=(5.7,5), fontsize=15)


plt.annotate('Equilibrium',xy=(7.3,7), xytext=(10.1, 7.5), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8))

plt.show()
```
