```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.facecolor'] = 'white'
```

```python
def demand_function1(p1, a, b, m, p2):
  """Calculates the demand for good 1 given its price, utility function parameters, income, and price of good 2.

  Args:
    p1: Price of good 1
    a: Coefficient of good X in the utility function.
    b: Coefficient of good Y in the utility function.
    m: Income.
    p2: Price of good 2

  Returns:
    Demand for good 1
  """

  if p1 / p2 > a / b:
    return 0
  else:
    return m / p1
  
def demand_function2(p2, a, b, m, p1):
  """Calculates the demand for good 2 given its price, utility function parameters, income, and price of good 1.

  Args:
    p2: Price of good 2
    a: Coefficient of good X in the utility function.
    b: Coefficient of good Y in the utility function.
    m: Income.
    p1: Price of good 1

  Returns:
    Demand for good 2
  """

  if p1 / p2 < a / b:
    return 0
  else:
    return m / p2
  
def demand_function_apt(p2, m):
  """demand for good 2 (apt-land), where apt-dev has no utility from house-land

  Args:
    p2: Price of good 2
    m: Income.

  Returns:
    Demand for good 2
  """
  return m / p2
  
# Linear supply function 

def supply_function(p,c1,c2):
  return c1+c2*p
```

```python
# n=2 individuals: house-dev and apt-dev
# shift supply of x2 right (apts), shift supply of x1 (houses) left
    # reallocating fixed stock of land S

# H-dev: u1(x1,x2) = x1 + x2
# A-dev: u2(x1,x2) = x2
a1,b1 = 1,1
m1,m2 = 24,12

# equilibrium 1
p1_eqm1, p2_eqm1 = 2,12

# equilibrium 2: 
p1_eqm2, p2_eqm2 = 24/11,6

# equilibrium 2: 
p1_eqm3, p2_eqm3 = 36/13, 36/13

# eqm 4 has same prices as eqm 3
```

```python
# demand for x1 (H)
# A-dev has no demand for H-land
p1_range = np.linspace(0.2, 15, 1000)

d11_eqm1 = [demand_function1(p1,a1,b1,m1,p2_eqm1) for p1 in p1_range]
d11_eqm2 = [demand_function1(p1,a1,b1,m1,p2_eqm2) for p1 in p1_range]
d11_eqm3 = [demand_function1(p1,a1,b1,m1,p2_eqm3) for p1 in p1_range]

D1_eqm1 = [x11 for x11 in zip(d11_eqm1)]
D1_eqm2 = [x11 for x11 in zip(d11_eqm2)]
D1_eqm3 = [x11 for x11 in zip(d11_eqm3)]
S1_eqm1 = [supply_function(p1,12,0) for p1 in p1_range]
S1_eqm2 = [supply_function(p1,11,0) for p1 in p1_range]
S1_eqm3 = [supply_function(p1,6.5,0) for p1 in p1_range]
S1_eqm4 = [supply_function(p1,0.1,0) for p1 in p1_range]

```

```python
# individual demand curves: x1

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(d11_eqm1, p1_range, label = "House developers", lw=4)
plt.xlabel("Quantity of house-zoned land")
plt.ylabel("Price of house-zoned land")
plt.title("Aggregate demand curve: house-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()
plt.show()
```

```python
# Eqm 1: aggregate demand and supply for x1

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D1_eqm1, p1_range, label="Demand", lw=4)
plt.plot(S1_eqm1, p1_range, label="Supply", lw=4)
plt.xlabel("Quantity of house-zoned land")
plt.ylabel("Price of house-zoned land")
plt.title("Supply and demand equilibrium: house-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()

plt.annotate('D',xy=(18,1.6), fontsize=15)
plt.annotate('S',xy=(12.3,13.5), fontsize=15)

plt.annotate('Equilibrium',xy=(12.2,2.1), xytext=(15.1,4.05), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# eqm 2: aggregate demand and supply for x1, 

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D1_eqm1, p1_range, label="Demand curve, initial", lw=4, alpha=0.5)
plt.plot(S1_eqm1, p1_range, label="Supply curve, initial", lw=4, alpha=0.5)
plt.plot(S1_eqm2, p1_range, label="Supply curve, final", lw=4)
plt.plot(D1_eqm2, p1_range, label="Demand curve, final", lw=4)
plt.xlabel("Quantity of house-zoned land")
plt.ylabel("Price of house-zoned land")
plt.title("Effect of a small upzoning: house-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()

# demand shift arrows
plt.annotate('',xy=(2.6,6.2), xytext=(2.6,8.7),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))
# supply shift arrows
plt.annotate('',xy=(10.7,15.3), xytext=(12.2, 15.3),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D1',xy=(3,9.1), fontsize=15)
plt.annotate('D2',xy=(3.5,4.7), fontsize=15)
plt.annotate('S1',xy=(12.3,13.5), fontsize=15)
plt.annotate('S2',xy=(9.7,13.5), fontsize=15)

plt.annotate('Old equilibrium',xy=(12.1,2.05), xytext=(13.5,4.05), fontsize=14,
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8, alpha=0.4), alpha=0.4)

plt.annotate('New equilibrium',xy=(10.8,2.1), xytext=(2.1,1.35), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# eqm 3: aggregate demand and supply for x1, 

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D1_eqm1, p1_range, label="Demand curve, initial", lw=4, alpha=0.5)
plt.plot(S1_eqm1, p1_range, label="Supply curve, initial", lw=4, alpha=0.5)
plt.plot(S1_eqm3, p1_range, label="Supply curve, final", lw=4)
plt.plot(D1_eqm3, p1_range, label="Demand curve, final", lw=4)
plt.xlabel("Quantity of house-zoned land")
plt.ylabel("Price of house-zoned land")
plt.title("Effect of a large upzoning: house-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()

# demand shift arrows
plt.annotate('',xy=(3,3), xytext=(3,7.6),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))
# supply shift arrows
plt.annotate('',xy=(6.8,13), xytext=(11.8, 13),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D1',xy=(4,6.1), fontsize=15)
plt.annotate('D2',xy=(4,3), fontsize=15)
plt.annotate('S1',xy=(12.3,13.5), fontsize=15)
plt.annotate('S2',xy=(6.7,13.5), fontsize=15)

plt.annotate('Old equilibrium',xy=(12.1,2.1), xytext=(13.5,4.05), fontsize=14,
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8, alpha=0.4), alpha=0.4)

plt.annotate('New equilibrium',xy=(6.3,2.6), xytext=(1,1.35), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))

plt.show()
```

```python
# eqm 4: aggregate demand and supply for x1, 

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D1_eqm1, p1_range, label="Demand curve, initial", lw=4, alpha=0.5)
plt.plot(S1_eqm1, p1_range, label="Supply curve, initial", lw=4, alpha=0.5)
plt.plot(S1_eqm4, p1_range, label="Supply curve, final", lw=4)
plt.plot(D1_eqm3, p1_range, label="Demand curve, final", lw=4)
plt.xlabel("Quantity of house-zoned land")
plt.ylabel("Price of house-zoned land")
plt.title("Effect of maximum upzoning: house-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()

# demand shift arrows
plt.annotate('',xy=(4,3), xytext=(4,5.7),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))
# supply shift arrows
plt.annotate('',xy=(0.2,13), xytext=(11.8, 13),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D1',xy=(4,6.1), fontsize=15)
plt.annotate('D2',xy=(4.5,3), fontsize=15)
plt.annotate('S1',xy=(12.3,13.5), fontsize=15)
plt.annotate('S2',xy=(0.3,13.5), fontsize=15)

plt.annotate('Old equilibrium',xy=(12.1,2.1), xytext=(13.5,4.05), fontsize=14,
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8, alpha=0.4), alpha=0.4)

plt.annotate('New equilibrium',xy=(0.3,2.6), xytext=(2.1,1.35), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# demand for x2 (apartment-zoned land)
p2_range = np.linspace(0.1, 15, 1000)

d12_eqm1 = [demand_function2(p2,a1,b1,m1,p1_eqm1) for p2 in p2_range]
d22_eqm1 = [demand_function_apt(p2,m2) for p2 in p2_range]

d12_eqm2 = [demand_function2(p2,a1,b1,m1,p1_eqm2) for p2 in p2_range]
d22_eqm2 = [demand_function_apt(p2,m2) for p2 in p2_range]

d12_eqm3 = [demand_function2(p2,a1,b1,m1,p1_eqm3) for p2 in p2_range]
d22_eqm3 = [demand_function_apt(p2,m2) for p2 in p2_range]


D2_eqm1 = [x12+x22 for x12,x22 in zip(d12_eqm1,d22_eqm1)]
D2_eqm2 = [x12+x22 for x12,x22 in zip(d12_eqm2,d22_eqm2)]
D2_eqm3 = [x12+x22 for x12,x22 in zip(d12_eqm3,d22_eqm3)]

S2_eqm1 = [supply_function(p2,1,0) for p2 in p2_range]
S2_eqm2 = [supply_function(p2,2,0) for p2 in p2_range]
S2_eqm3 = [supply_function(p2,6.5,0) for p2 in p2_range]
S2_eqm4 = [supply_function(p2,13,0) for p2 in p2_range]
```

```python
# individual demand curves: x2

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(d12_eqm1, p2_range, label = "House developers", lw=4)
plt.plot(d22_eqm1, p2_range, label = "Apartment developers", lw=4)
plt.xlabel("Quantity of apartment-zoned land")
plt.ylabel("Price of apartment-zoned land")
plt.title("Sector demand curves: apartment-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()
plt.show()
```

```python
# aggregate demand for x2
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve", lw=4)
plt.xlabel("Quantity of apartment-zoned land")
plt.ylabel("Price of apartment-zoned land")
plt.title("Aggregate demand curve: apartment-zoned land")
plt.grid(True)
ax.set_xlim(0, 30)
plt.legend()

plt.annotate('Apartment developers buy',xy=(1.4,10), xytext=(6, 10.5), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))
plt.annotate('Both Apartment & House developers buy',xy=(26,1.5), xytext=(12,6.1), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))
plt.annotate('',xy=(12,2.2), xytext=(17,5.9), fontsize=14, 
        arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# eqm1: aggregate supply and demand for x2
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve", lw=4)
plt.plot(S2_eqm1, p2_range, label="Supply curve", lw=4)
plt.xlabel("Quantity of apartment-zoned land")
plt.ylabel("Price of apartment-zoned land")
plt.title("Initial equilibrium: apartment-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()

# label curves
plt.annotate('D',xy=(11.5,2.2), fontsize=15)
plt.annotate('S',xy=(1.2,3.2), fontsize=15)

plt.annotate('Equilibrium',xy=(1.15,12), xytext=(6, 10.5), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8))


plt.show()
```

```python
# Eqm 2: small upzoning: shift supply for x2

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve", lw=4, alpha=0.5)
plt.plot(S2_eqm1, p2_range, label="Supply curve, initial", lw=4, alpha=0.5)
plt.plot(S2_eqm2, p2_range, label="Supply curve, final", lw=4)
plt.plot(D2_eqm2, p2_range, label="Demand curve, final", lw=4)
plt.xlabel("Quantity of apartment-zoned land")
plt.ylabel("Price of apartment-zoned land")
plt.title("Effect of small upzoning: apartment-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()

# supply shift arrow
plt.annotate('',xy=(2.3, -0.2), xytext=(0.8,-0.2),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D1',xy=(11.5,1.3), fontsize=15)
plt.annotate('D2',xy=(11.5,2.5), fontsize=15)
plt.annotate('S1',xy=(0.1,1.1), fontsize=15)
plt.annotate('S2',xy=(2.2,1.1), fontsize=15)

plt.annotate('Old equilibrium',xy=(1.15,12), xytext=(6, 10.5), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8, alpha=0.4), alpha=0.4)
plt.annotate('New equilibrium',xy=(2.2,6.1), xytext=(5.1,7), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8))

plt.show()
```

```python
# Eqm 3: big upzoning: shift supply for x2

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve", lw=4, alpha=0.5)
plt.plot(S2_eqm1, p2_range, label="Supply curve, initial", lw=4, alpha=0.5)
plt.plot(S2_eqm3, p2_range, label="Supply curve, final", lw=4)
plt.plot(D2_eqm3, p2_range, label="Demand curve, final", lw=4)
plt.xlabel("Quantity of apartment-zoned land")
plt.ylabel("Price of apartment-zoned land")
plt.title("Effect of large upzoning: apartment-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()

# demand shift arrows
plt.annotate('',xy=(11,2.8), xytext=(11,2),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))
# supply shift arrow
plt.annotate('',xy=(6.3, 0.7), xytext=(1.2,0.7),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D1',xy=(11.5,1.3), fontsize=15)
plt.annotate('D2',xy=(11.5,3), fontsize=15)
plt.annotate('S1',xy=(0.1,1.1), fontsize=15)
plt.annotate('S2',xy=(5.5,1.1), fontsize=15)

plt.annotate('Old equilibrium',xy=(1.15,12), xytext=(2, 10.5), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8, alpha=0.4), alpha=0.4)
plt.annotate('New equilibrium',xy=(6.7,3), xytext=(8.7, 4.1), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8))

plt.show()
```

```python
# Eqm 4: max upzoning: shift supply for x2

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plt.plot(D2_eqm1, p2_range, label="Demand curve", lw=4, alpha=0.5)
plt.plot(S2_eqm1, p2_range, label="Supply curve, initial", lw=4, alpha=0.5)
plt.plot(S2_eqm4, p2_range, label="Supply curve, final", lw=4,)
plt.plot(D2_eqm3, p2_range, label="Demand curve, final", lw=4)
plt.xlabel("Quantity of apartment-zoned land")
plt.ylabel("Price of apartment-zoned land")
plt.title("Effect of maximum upzoning: apartment-zoned land")
plt.grid(True)
ax.set_xlim(0, 25)
plt.legend()

# demand shift arrows
plt.annotate('',xy=(9.4,2.8), xytext=(9.4,2),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))
# supply shift arrow
plt.annotate('',xy=(12.6,0.7), xytext=(1.4, 0.7),arrowprops=dict(
    facecolor='black', width=0.5, headwidth=8))

# label curves
plt.annotate('D1',xy=(8,1.3), fontsize=15)
plt.annotate('D2',xy=(8,3), fontsize=15)
plt.annotate('S1',xy=(0.1,1.1), fontsize=15)
plt.annotate('S2',xy=(11.8,1.1), fontsize=15)

plt.annotate('Old equilibrium',xy=(1.15,12), xytext=(2, 10.5), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8, alpha=0.4), alpha=0.4)
plt.annotate('New equilibrium',xy=(13.2,2.9), xytext=(15.6, 4.1), fontsize=14,
             arrowprops=dict(facecolor='black', width=0.5, headwidth=8))

plt.show()
```
